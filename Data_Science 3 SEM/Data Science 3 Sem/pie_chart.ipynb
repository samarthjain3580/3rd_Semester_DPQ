```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [10,20,30,40,50]
label = ['I','II','III','IV','V']
c=['red','pink','yellow','cyan','orange']
plt.pie(data, labels=label, colors=c)
plt.show()
```

```python
data_2 = [10,20,30,40,50]
label = ['I','II','III','IV','V']
c=['red','pink','yellow','cyan','orange']
plt.pie(data_2, labels=label, colors=c, startangle=90, explode=[0.2,0,0,0,0.1])
plt.legend(title='5 Categories')
plt.show()
```

```python
data_3 = [10,20,30,40,50]
label = ['I','II','III','IV','V']
c=['red','pink','yellow','cyan','orange']
plt.pie(data_3, labels=label, colors=c,shadow = True, autopct='%.1f%%')
plt.legend(title='5 Categories')
plt.show()
```
