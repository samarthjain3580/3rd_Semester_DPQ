```python
x = np.linspace(0,4*np.pi,100)
y = np.cos(x)
```

```python
plt.plot(x,y, marker = "+")             #marker: (D)-diamond, (H)-Hexagon, can also be used operators
plt.show()
```

```python
plt.plot(x,y,ls="--", color = "fuchsia", marker = "+", ms="2", mec="r", linewidth = 3)   #Attribute : ls(linestyle) , color
plt.xlabel("Angle")
plt.ylabel("Value of Cos(x)")   
plt.title("Cosine Graph")
plt.grid(ls="dashed", color = "blue")                                                      #axis(x)/axis(y) can also be given
plt.show()       
```

```python
plt.suptitle("Earth")
plt.subplot(1,2,1)          #subplot(row, column, position)
plt.title("USA")
plt.subplot(1,2,2)
plt.title("India")
plt.show()
```
